```python
import plotly as py
import numpy as np
import plotly.express as px
import plotly.graph_objs as go
import pandas as pd
import cufflinks as cf

%matplotlib inline
cf.go_offline()

df = pd.read_csv('Artificial - Loan Process.csv')
```

```python
fig = px.scatter(df, x=df.iloc[:,2].values.tolist(), y=df.iloc[:,0].values.tolist(), color=df.iloc[:,1].values.tolist(), 
                labels={
                    'x':df.columns[2],
                    'y':df.columns[0],
                    'color':df.columns[1],
                },
                title= "Traces over Time"
                )
fig.show()
```
