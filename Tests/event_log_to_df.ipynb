## Converting XES to pandas data frame

```python
from pm4py.objects.log.importer.xes import importer as xes_importer
import pandas as pd
from pm4py.objects.conversion.log import converter as log_converter

variant = xes_importer.Variants.ITERPARSE
parameters = {variant.value.Parameters.TIMESTAMP_SORT: True}
log = xes_importer.apply('running-example.xes',
                         variant=variant, parameters=parameters)
#parameters = {log_converter.Variants.TO_DATA_FRAME.value.Parameters.CASE_ID_KEY: 'case:'}
df_event_log = log_converter.apply(log, variant=log_converter.Variants.TO_DATA_FRAME)
df_event_log
```
